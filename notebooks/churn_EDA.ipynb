## Feature Enginnering

```python
# convert totla charges to numeric columns

churn_data["TotalCharges"] = pd.to_numeric(churn_data["TotalCharges"],errors="coerce")
```

```python
churn_data["TotalCharges"].dtype
```

```python
churn_data["TotalCharges"].isnull().sum()
```

```python
# handle null values in Total charges

churn_data["TotalCharges"] = churn_data["TotalCharges"].fillna(churn_data["TotalCharges"].median())
churn_data["TotalCharges"].isnull().sum()
```

```python
# Create new features

churn_data["AvgChargesPerMonth"] = churn_data["TotalCharges"]/ (churn_data["tenure"]+1)
```

```python
# separate target

y = churn_data["Churn"].map({"Yes" :1, "No" :0})
X = churn_data.drop("Churn",axis =1)
```

```python
# one hot encode for categorical columns
cate_col = churn_data.select_dtypes(include="object").columns
churn_data= pd.get_dummies(churn_data,columns= cate_col,drop_first=True)
```
